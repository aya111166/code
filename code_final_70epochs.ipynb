```python
from __future__ import absolute_import, division, print_function, unicode_literals

import tensorflow as tf
from tensorflow.keras import models , layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D,BatchNormalization ,Flatten ,Reshape  ,RepeatVector ,Concatenate
from tensorflow.keras.layers import  Conv2D,LSTM ,Bidirectional  ,Input,Embedding  ,GlobalMaxPool1D ,TimeDistributed

from tensorflow.keras.layers import Activation,MaxPooling2D,AveragePooling3D,GlobalMaxPool2D,AveragePooling2D
from tensorflow.keras.layers import MaxPooling3D,GlobalAveragePooling3D,GlobalMaxPool3D,Conv3D

from sklearn.model_selection import train_test_split ,StratifiedKFold ,KFold ,RepeatedKFold
from sklearn import preprocessing

from tensorflow.keras import Model 

from keras.utils import np_utils

from sklearn.cluster import KMeans
from sklearn import cluster


import matplotlib.pyplot as plt
import numpy as np
import scipy.io
import os

import pylab as pl

from scipy import interp
from itertools import cycle
import matplotlib.patches as patches


from sklearn.metrics import accuracy_score ,roc_curve, auc ,roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


def invert_permutation(p):
    '''The argument p is assumed to be some permutation of 0, 1, ..., len(p)-1. 
    Returns an array s, where s[i] gives the index of i in p.
    '''
    s = np.empty(p.size, p.dtype)
    s[p] = np.arange(p.size)
    return s


from tensorflow.keras import backend as K
def pearson_r(y_true, y_pred):
    x = y_true
    y = y_pred
    mx = K.mean(x, axis=0)
    my = K.mean(y, axis=0)
    xm, ym = x - mx, y - my
    r_num = K.sum(xm * ym)
    x_square_sum = K.sum(xm * xm)
    y_square_sum = K.sum(ym * ym)
    r_den = K.sqrt(x_square_sum * y_square_sum)
    r = r_num / r_den
    return K.mean(r)



import os 
from PIL import Image
import numpy as np
from numpy import *   

from matplotlib import pyplot as plt
from IPython import display


from patchify import patchify, unpatchify


import matplotlib.pyplot as plt
%pylab inline
%load_ext tensorboard
```

```python
def cancer_NC(splits , epochs,batch_size):


    mat_sc= scipy.io.loadmat('cancer_img_p')
    cancer_img_p = mat_sc['cancer_img_p']

    mat_sc= scipy.io.loadmat('normal_img_p')
    normal_img_p = mat_sc['normal_img_p']

    labels_cancer_ = np.ones((cancer_img_p.shape[0],1))
    labels_normal_ = np.zeros((normal_img_p.shape[0],1))



    x_cancer_NC_ = np.concatenate((cancer_img_p, normal_img_p), axis=0) 
    y_cancer_NC_ = np.concatenate((labels_cancer_, labels_normal_), axis=0) 

    x_cancer_NC_=np.asarray(tf.expand_dims(x_cancer_NC_, axis=-1))





    x_cancer_NC_1 = x_cancer_NC_[:,0,0]
    x_cancer_NC_2 = x_cancer_NC_[:,1,0]
    x_cancer_NC_3 = x_cancer_NC_[:,2,0]
    x_cancer_NC_4 = x_cancer_NC_[:,3,0]
    x_cancer_NC_5 = x_cancer_NC_[:,4,0]
    x_cancer_NC_6 = x_cancer_NC_[:,5,0]

    x_cancer_NC_7 = x_cancer_NC_[:,6,0]
    x_cancer_NC_8 = x_cancer_NC_[:,7,0]
    x_cancer_NC_9 = x_cancer_NC_[:,8,0]
    x_cancer_NC_10 = x_cancer_NC_[:,9,0]
    x_cancer_NC_11 = x_cancer_NC_[:,10,0]




    n = x_cancer_NC_1.shape[0]
    permutation = list(np.random.permutation(n))
    x_cancer_NC_1 = x_cancer_NC_1[permutation]
    x_cancer_NC_2 = x_cancer_NC_2[permutation]
    x_cancer_NC_3 = x_cancer_NC_3[permutation]
    x_cancer_NC_4 = x_cancer_NC_4[permutation]
    x_cancer_NC_5 = x_cancer_NC_5[permutation]
    x_cancer_NC_6 = x_cancer_NC_6[permutation]


    x_cancer_NC_7 = x_cancer_NC_7[permutation]
    x_cancer_NC_8 = x_cancer_NC_8[permutation]
    x_cancer_NC_9 = x_cancer_NC_9[permutation]
    x_cancer_NC_10 = x_cancer_NC_10[permutation]
    x_cancer_NC_11 = x_cancer_NC_11[permutation]




    y_cancer_NC_ = y_cancer_NC_[permutation]

    permutation_xray = np.asarray(permutation)




    input_1 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_1')



    conv_1 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_1)
    print(conv_1.shape)
    conv_1 = BatchNormalization()(conv_1)
    conv_1 = Dropout(rate=0.5)(conv_1)


    conv_1 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_1)
    print(conv_1.shape)
    conv_1 = BatchNormalization()(conv_1)
    conv_1 = Dropout(rate=0.5)(conv_1)



    conv_1 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_1)
    print(conv_1.shape)
    conv_1 = BatchNormalization()(conv_1)
    conv_1 = Dropout(rate=0.5)(conv_1)


    conv_1 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_1)
    print(conv_1.shape)
    conv_1 = BatchNormalization()(conv_1)
    conv_1 = Dropout(rate=0.5)(conv_1)


    conv_1 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_1)
    print(conv_1.shape)
    conv_1 = BatchNormalization()(conv_1)
    conv_1 = Dropout(rate=0.5)(conv_1)


    conv_1 = AveragePooling2D()(conv_1)












    input_2 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_2')

    conv_2 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_2)
    print(conv_2.shape)
    conv_2 = BatchNormalization()(conv_2)
    conv_2 = Dropout(rate=0.5)(conv_2)


    conv_2 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_2)
    print(conv_2.shape)
    conv_2 = BatchNormalization()(conv_2)
    conv_2 = Dropout(rate=0.5)(conv_2)



    conv_2 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_2)
    print(conv_2.shape)
    conv_2 = BatchNormalization()(conv_2)
    conv_2 = Dropout(rate=0.5)(conv_2)


    conv_2 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_2)
    print(conv_2.shape)
    conv_2 = BatchNormalization()(conv_2)
    conv_2 = Dropout(rate=0.5)(conv_2)


    conv_2 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_2)
    print(conv_2.shape)
    conv_2 = BatchNormalization()(conv_2)
    conv_2 = Dropout(rate=0.5)(conv_2)


    conv_2 = AveragePooling2D()(conv_2)







    input_3= tf.keras.layers.Input(shape=[256, 256,1], name='main_input_3')

    conv_3 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_3)
    print(conv_3.shape)
    conv_3 = BatchNormalization()(conv_3)
    conv_3 = Dropout(rate=0.5)(conv_3)


    conv_3 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_3)
    print(conv_3.shape)
    conv_3 = BatchNormalization()(conv_3)
    conv_3 = Dropout(rate=0.5)(conv_3)



    conv_3 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_3)
    print(conv_3.shape)
    conv_3 = BatchNormalization()(conv_3)
    conv_3 = Dropout(rate=0.5)(conv_3)


    conv_3 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_3)
    print(conv_3.shape)
    conv_3 = BatchNormalization()(conv_3)
    conv_3 = Dropout(rate=0.5)(conv_3)


    conv_3 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_3)
    print(conv_3.shape)
    conv_3 = BatchNormalization()(conv_3)
    conv_3 = Dropout(rate=0.5)(conv_3)


    conv_3 = AveragePooling2D()(conv_3)





    input_4= tf.keras.layers.Input(shape=[256, 256,1], name='main_input_4')


    conv_4 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_4)
    print(conv_4.shape)
    conv_4 = BatchNormalization()(conv_4)
    conv_4 = Dropout(rate=0.5)(conv_4)


    conv_4 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_4)
    print(conv_4.shape)
    conv_4 = BatchNormalization()(conv_4)
    conv_4 = Dropout(rate=0.5)(conv_4)



    conv_4 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_4)
    print(conv_4.shape)
    conv_4 = BatchNormalization()(conv_4)
    conv_4 = Dropout(rate=0.5)(conv_4)


    conv_4 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_4)
    print(conv_4.shape)
    conv_4 = BatchNormalization()(conv_4)
    conv_4 = Dropout(rate=0.5)(conv_4)


    conv_4 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_4)
    print(conv_4.shape)
    conv_4 = BatchNormalization()(conv_4)
    conv_4 = Dropout(rate=0.5)(conv_4)


    conv_4 = AveragePooling2D()(conv_4)







    input_5 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_5')

    conv_5 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_5)
    print(conv_5.shape)
    conv_5 = BatchNormalization()(conv_5)
    conv_5 = Dropout(rate=0.5)(conv_5)


    conv_5 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_5)
    print(conv_5.shape)
    conv_5 = BatchNormalization()(conv_5)
    conv_5 = Dropout(rate=0.5)(conv_5)



    conv_5 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_5)
    print(conv_5.shape)
    conv_5 = BatchNormalization()(conv_5)
    conv_5 = Dropout(rate=0.5)(conv_5)


    conv_5 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_5)
    print(conv_5.shape)
    conv_5 = BatchNormalization()(conv_5)
    conv_5 = Dropout(rate=0.5)(conv_5)


    conv_5 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_5)
    print(conv_5.shape)
    conv_5 = BatchNormalization()(conv_5)
    conv_5 = Dropout(rate=0.5)(conv_5)


    conv_5 = AveragePooling2D()(conv_5)





    input_6 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_6')


    conv_6 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_6)
    print(conv_6.shape)
    conv_6 = BatchNormalization()(conv_6)
    conv_6 = Dropout(rate=0.5)(conv_6)


    conv_6 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_6)
    print(conv_6.shape)
    conv_6 = BatchNormalization()(conv_6)
    conv_6 = Dropout(rate=0.5)(conv_6)



    conv_6 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_6)
    print(conv_6.shape)
    conv_6 = BatchNormalization()(conv_6)
    conv_6 = Dropout(rate=0.5)(conv_6)


    conv_6 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_6)
    print(conv_6.shape)
    conv_6 = BatchNormalization()(conv_6)
    conv_6 = Dropout(rate=0.5)(conv_6)


    conv_6 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_6)
    print(conv_6.shape)
    conv_6 = BatchNormalization()(conv_6)
    conv_6 = Dropout(rate=0.5)(conv_6)


    conv_6 = AveragePooling2D()(conv_6)




    input_7 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_7')



    conv_7 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_7)
    print(conv_7.shape)
    conv_7 = BatchNormalization()(conv_7)
    conv_7 = Dropout(rate=0.5)(conv_7)


    conv_7 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_7)
    print(conv_7.shape)
    conv_7 = BatchNormalization()(conv_7)
    conv_7 = Dropout(rate=0.5)(conv_7)



    conv_7 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_7)
    print(conv_7.shape)
    conv_7 = BatchNormalization()(conv_7)
    conv_7 = Dropout(rate=0.5)(conv_7)


    conv_7 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_7)
    print(conv_7.shape)
    conv_7 = BatchNormalization()(conv_7)
    conv_7 = Dropout(rate=0.5)(conv_7)


    conv_7 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_7)
    print(conv_7.shape)
    conv_7 = BatchNormalization()(conv_7)
    conv_7 = Dropout(rate=0.5)(conv_7)


    conv_7 = AveragePooling2D()(conv_7)







    input_8 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_8')



    conv_8 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_8)
    print(conv_8.shape)
    conv_8 = BatchNormalization()(conv_8)
    conv_8 = Dropout(rate=0.5)(conv_8)


    conv_8 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_8)
    print(conv_8.shape)
    conv_8 = BatchNormalization()(conv_8)
    conv_8 = Dropout(rate=0.5)(conv_8)



    conv_8 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_8)
    print(conv_8.shape)
    conv_8 = BatchNormalization()(conv_8)
    conv_8 = Dropout(rate=0.5)(conv_8)


    conv_8 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_8)
    print(conv_8.shape)
    conv_8 = BatchNormalization()(conv_8)
    conv_8 = Dropout(rate=0.5)(conv_8)


    conv_8 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_8)
    print(conv_8.shape)
    conv_8 = BatchNormalization()(conv_8)
    conv_8 = Dropout(rate=0.5)(conv_8)


    conv_8 = AveragePooling2D()(conv_8)






    input_9 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_9')



    conv_9 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_9)
    print(conv_9.shape)
    conv_9 = BatchNormalization()(conv_9)
    conv_9 = Dropout(rate=0.5)(conv_9)


    conv_9 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_9)
    print(conv_9.shape)
    conv_9 = BatchNormalization()(conv_9)
    conv_9 = Dropout(rate=0.5)(conv_9)



    conv_9 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_9)
    print(conv_9.shape)
    conv_9 = BatchNormalization()(conv_9)
    conv_9 = Dropout(rate=0.5)(conv_9)


    conv_9 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_9)
    print(conv_9.shape)
    conv_9 = BatchNormalization()(conv_9)
    conv_9 = Dropout(rate=0.5)(conv_9)


    conv_9 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_9)
    print(conv_9.shape)
    conv_9 = BatchNormalization()(conv_9)
    conv_9 = Dropout(rate=0.5)(conv_9)


    conv_9 = AveragePooling2D()(conv_9)





    input_10 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_10')



    conv_10 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_10)
    print(conv_10.shape)
    conv_10 = BatchNormalization()(conv_10)
    conv_10 = Dropout(rate=0.5)(conv_10)


    conv_10 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_10)
    print(conv_10.shape)
    conv_10 = BatchNormalization()(conv_10)
    conv_10 = Dropout(rate=0.5)(conv_10)



    conv_10 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_10)
    print(conv_10.shape)
    conv_10 = BatchNormalization()(conv_10)
    conv_10 = Dropout(rate=0.5)(conv_10)


    conv_10 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_10)
    print(conv_10.shape)
    conv_10 = BatchNormalization()(conv_10)
    conv_10 = Dropout(rate=0.5)(conv_10)


    conv_10 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_10)
    print(conv_10.shape)
    conv_10 = BatchNormalization()(conv_10)
    conv_10 = Dropout(rate=0.5)(conv_10)


    conv_10 = AveragePooling2D()(conv_10)






    input_11 = tf.keras.layers.Input(shape=[256, 256,1], name='main_input_11')



    conv_11 = layers.Conv2D(32, (2, 2), strides=(2, 2),activation="relu")(input_11)
    print(conv_11.shape)
    conv_11 = BatchNormalization()(conv_11)
    conv_11 = Dropout(rate=0.5)(conv_11)


    conv_11 = layers.Conv2D(64, (2, 2), strides=(2, 2),activation="relu" )(conv_11)
    print(conv_11.shape)
    conv_11 = BatchNormalization()(conv_11)
    conv_11 = Dropout(rate=0.5)(conv_11)



    conv_11 = layers.Conv2D(128, (2, 2), strides=(2, 2),activation="relu" )(conv_11)
    print(conv_11.shape)
    conv_11 = BatchNormalization()(conv_11)
    conv_11 = Dropout(rate=0.5)(conv_11)


    conv_11 = layers.Conv2D(256, (2, 2), strides=(2, 2),activation="relu" )(conv_11)
    print(conv_11.shape)
    conv_11 = BatchNormalization()(conv_11)
    conv_11 = Dropout(rate=0.5)(conv_11)


    conv_11 = layers.Conv2D(512, (2, 2), strides=(2, 2),activation="relu" )(conv_11)
    print(conv_11.shape)
    conv_11 = BatchNormalization()(conv_11)
    conv_11 = Dropout(rate=0.5)(conv_11)


    conv_11 = AveragePooling2D()(conv_11)











    x = Concatenate(axis =-1)([conv_1,conv_2])
    x = Concatenate(axis =-1)([x,conv_3])
    x = Concatenate(axis =-1)([x,conv_4])
    x = Concatenate(axis =-1)([x,conv_5])
    x = Concatenate(axis =-1)([x,conv_6])

    x = Concatenate(axis =-1)([x,conv_7])
    x = Concatenate(axis =-1)([x,conv_8])
    x = Concatenate(axis =-1)([x,conv_9])
    x = Concatenate(axis =-1)([x,conv_10])
    x = Concatenate(axis =-1)([x,conv_11])






    x_4 = tf.keras.layers.Flatten()(x)
    print(x_4.shape)

    x_4 = tf.keras.layers.Dense(64,  name="layer_dense")(x_4)

    x_4 = tf.keras.layers.Dense(2, activation='softmax' , name="layer_x")(x_4)


    adam = tf.keras.optimizers.Adam( lr=1e-4)

    model = tf.keras.Model(inputs= [input_1,input_2,input_3,input_4,input_5,input_6,
                                   input_7,input_8,input_9,input_10,input_11], outputs= x_4)

    model.compile(optimizer= adam , 
                      loss= { 'layer_x':'binary_crossentropy'},
                      metrics={ 'layer_x':'accuracy'})




    kfold = StratifiedKFold(n_splits=splits,  random_state=np.random.seed(0))


    accuracy_train_x_=[]
    accuracy_test_x_=[] 

    mean_tpr = 0.0
    mean_fpr = np.linspace(0, 1, 100)
    all_tpr = []


    accuracy_tr_all=[]
    accuracy_te_all=[]

    confusion_matrix_1_last =[]
 

    count = 1
    for train_index, test_index in kfold.split(x_cancer_NC_1, y_cancer_NC_):
        print(count)
        count = count+1

        x_train_1, x_test_1 = x_cancer_NC_1[train_index], x_cancer_NC_1[test_index]
        y_train, y_test = y_cancer_NC_[train_index], y_cancer_NC_[test_index]
        
        

        x_train_2, x_test_2 = x_cancer_NC_2[train_index], x_cancer_NC_2[test_index]
        x_train_3, x_test_3 = x_cancer_NC_3[train_index], x_cancer_NC_3[test_index]
        x_train_4, x_test_4 = x_cancer_NC_4[train_index], x_cancer_NC_4[test_index]
        x_train_5, x_test_5 = x_cancer_NC_5[train_index], x_cancer_NC_5[test_index]
        x_train_6, x_test_6 = x_cancer_NC_6[train_index], x_cancer_NC_6[test_index]

        
        x_train_7, x_test_7 = x_cancer_NC_7[train_index], x_cancer_NC_7[test_index]
        x_train_8, x_test_8 = x_cancer_NC_8[train_index], x_cancer_NC_8[test_index]
        x_train_9, x_test_9 = x_cancer_NC_9[train_index], x_cancer_NC_9[test_index]
        x_train_10, x_test_10 = x_cancer_NC_10[train_index], x_cancer_NC_10[test_index]
        x_train_11, x_test_11 = x_cancer_NC_11[train_index], x_cancer_NC_11[test_index]

             

        ## normalization : normalize each feature for norm = 1
        # axis 0 for columns ( features) and axis 1 for rows for this function
        # f.keras.utils.normalize() and also this function make l2 norm according to the chosen axis

        x_train_1 = tf.keras.utils.normalize(x_train_1, axis=1)
        x_train_2 = tf.keras.utils.normalize(x_train_2, axis=1)
        x_train_3 = tf.keras.utils.normalize(x_train_3, axis=1)
        x_train_4 = tf.keras.utils.normalize(x_train_4, axis=1)
        x_train_5 = tf.keras.utils.normalize(x_train_5, axis=1)
        x_train_6 = tf.keras.utils.normalize(x_train_6, axis=1)
        
        x_train_7 = tf.keras.utils.normalize(x_train_7, axis=1)
        x_train_8 = tf.keras.utils.normalize(x_train_8, axis=1)
        x_train_9 = tf.keras.utils.normalize(x_train_9, axis=1)
        x_train_10 = tf.keras.utils.normalize(x_train_10, axis=1)
        x_train_11 = tf.keras.utils.normalize(x_train_11, axis=1)
        
        
       
        x_test_1 = tf.keras.utils.normalize(x_test_1, axis=1)
        x_test_2 = tf.keras.utils.normalize(x_test_2, axis=1)
        x_test_3 = tf.keras.utils.normalize(x_test_3, axis=1)
        x_test_4 = tf.keras.utils.normalize(x_test_4, axis=1)
        x_test_5 = tf.keras.utils.normalize(x_test_5, axis=1)
        x_test_6 = tf.keras.utils.normalize(x_test_6, axis=1)
        
        x_test_7 = tf.keras.utils.normalize(x_test_7, axis=1)
        x_test_8 = tf.keras.utils.normalize(x_test_8, axis=1)
        x_test_9 = tf.keras.utils.normalize(x_test_9, axis=1)
        x_test_10 = tf.keras.utils.normalize(x_test_10, axis=1)
        x_test_11 = tf.keras.utils.normalize(x_test_11, axis=1)
        
        
       
        ## two output for classification
        
    
        ## two output for classification

        uniques_y_test, ids_y_test = np.unique(y_test, return_inverse=True)   
        
        
        y_train = np_utils.to_categorical(y_train, num_classes=2)
        y_test  =  np_utils.to_categorical(y_test, num_classes=2)


        history = model.fit([x_train_1,x_train_2,x_train_3,x_train_4,x_train_5,x_train_6,
                            x_train_7,x_train_8,x_train_9,x_train_10,x_train_11], y_train, 
                            epochs=epochs, batch_size=batch_size , 
                            validation_data=( [x_test_1,x_test_2,x_test_3,x_test_4,x_test_5,x_test_6,
                                              x_test_7,x_test_8,x_test_9,x_test_10,x_test_11] , y_test))

        

        
        
        
        print(history.history.keys())
        a,accuracy_train_x = model.evaluate([x_train_1,x_train_2,x_train_3,x_train_4,x_train_5,x_train_6,
                                            x_train_7,x_train_8,x_train_9,x_train_10,x_train_11],y_train)

        a, accuracy_test_x = model.evaluate([x_test_1,x_test_2,x_test_3,x_test_4,x_test_5,x_test_6,
                                            x_test_7,x_test_8,x_test_9,x_test_10,x_test_11],y_test)

        accuracy_train_x_.append(accuracy_train_x)

        accuracy_test_x_.append(accuracy_test_x)   
   



        # classification paramaters


        yhat_probs = model.predict([x_test_1,x_test_2,x_test_3,x_test_4,x_test_5,x_test_6,
                                                      x_test_7,x_test_8,x_test_9,x_test_10,x_test_11])


        yhat_classes =yhat_probs>0.5
        yhat_classes_1 =  1*yhat_classes 


        # confusion matrix
        y_test_con = uniques_y_test[y_test.argmax(1)]
        yhat_classes_1_con = uniques_y_test[yhat_classes_1.argmax(1)]

        confusion_matrix_1_last.append(confusion_matrix(y_test_con, yhat_classes_1_con,labels=[0, 1]))


        # Compute ROC curve and area the curve
        fpr, tpr, thresholds = roc_curve( y_cancer_NC_[test_index], yhat_classes_1_con)  
        mean_tpr += interp(mean_fpr, fpr, tpr)
        mean_tpr[0] = 0.0
        roc_auc = auc(fpr, tpr)
        

    mean_tpr /= splits
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)

    

    
    
    return    accuracy_train_x_,accuracy_test_x_,confusion_matrix_1_last,mean_fpr,mean_tpr,mean_auc    



```
